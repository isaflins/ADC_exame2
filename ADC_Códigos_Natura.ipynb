<span style="font-size: 28px; color: Black; font-family: 'Times New Roman';"> **1. Indicadores de Liquidez e Atividade**</span>


<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$Liquidez\ Imediata=\frac{Disponível}{Passivo\ Circulante}$$</span>


```python
#Liquidez Imediata em 2024
LI_2024 = disponível_2024 / PC_2024

#Liquidez Imediata em 2023
LI_2023 = disponível_2023 / PC_2023

#Liquidez Imediata em 2022
LI_2022 = disponível_2022 / PC_2022

#Liquidez Imediata em 2021
LI_2021 = disponível_2021 / PC_2021

#Liquidez Imediata em 2020
LI_2020 = disponível_2020 / PC_2020

#DataFrame com os resultados
liquidez_imediata = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Liquidez Imediata': [LI_2020, LI_2021, LI_2022, LI_2023, LI_2024]
}

df_liquidez_imediata = pd.DataFrame(liquidez_imediata)

#Definindo o ano como índice
df_liquidez_imediata.set_index('Ano', inplace=True)

df_liquidez_imediata
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$Liquidez\ Corrente=\frac{Ativo\ Circulante}{Passivo\ Circulante}$$</span>


```python
#Liquidez Corrente em 2024
LC_2024 = AC_2024 / PC_2024

#Liquidez Corrente em 2023
LC_2023 = AC_2023 / PC_2023

#Liquidez Corrente em 2022
LC_2022 = AC_2022 / PC_2022

#Liquidez Corrente em 2021
LC_2021 = AC_2021 / PC_2021

#Liquidez Corrente em 2020
LC_2020 = AC_2020 / PC_2020

#DataFrame com os resultados
liquidez_corrente = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Liquidez Corrente': [LC_2020, LC_2021, LC_2022, LC_2023, LC_2024]
}

df_liquidez_corrente = pd.DataFrame(liquidez_corrente)

#Definindo o ano como índice
df_liquidez_corrente.set_index('Ano', inplace=True)

df_liquidez_corrente
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Liquidez Geral}=\frac{\text{Ativo Circulante}+\text{Realizável a Longo Prazo}}{\text{Passivo Circulante}+\text{Exigível a Longo Prazo}}$$</span>


```python
#Liquidez Geral em 2024
LG_2024 = (AC_2024 + RLP_2024) / (PC_2024 + PNC_2024) 

#Liquidez Geral em 2023
LG_2023 = (AC_2023 + RLP_2023) / (PC_2023 + PNC_2023) 

#Liquidez Geral em 2022
LG_2022 = (AC_2022 + RLP_2022) / (PC_2022 + PNC_2022)

#Liquidez Geral em 2021
LG_2021 = (AC_2021 + RLP_2021) / (PC_2021 + PNC_2021)

#Liquidez Geral em 2020
LG_2020 = (AC_2020 + RLP_2020) / (PC_2020 + PNC_2020)

#DataFrame com os resultados
liquidez_geral = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Liquidez Geral': [LG_2020, LG_2021, LG_2022, LG_2023, LG_2024]
}

df_liquidez_geral = pd.DataFrame(liquidez_geral)

#Definindo o ano como índice
df_liquidez_geral.set_index('Ano', inplace=False)
df_liquidez_geral['Ano'] = df_liquidez_geral['Ano'].astype(str)

df_liquidez_geral
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Capital de Giro}=(\text{PL}+\text{Exigível a Longo Prazo})-({\text{Ativo não Circulante})}$$</span>


```python
#Capital Circulante Líquido em 2024
CCL_2024 = (PL_2024 + PNC_2024) - (ANC_2024)

#Capital Circulante Líquido em 2023
CCL_2023 = (PL_2023 + PNC_2023) - (ANC_2023)

#Capital Circulante Líquido em 2022
CCL_2022 = (PL_2022 + PNC_2022) - (ANC_2022)

#Capital Circulante Líquido em 2021
CCL_2021 = (PL_2021 + PNC_2021) - (ANC_2021)

#Capital Circulante Líquido em 2020
CCL_2020 = (PL_2020 + PNC_2020) - (ANC_2020)

#DataFrame com os resultados
CCL = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Capital Circulante Líquido': [CCL_2020, CCL_2021, CCL_2022, CCL_2023, CCL_2024]
}

df_CCL = pd.DataFrame(CCL)

#Definindo o ano como índice
df_CCL.set_index('Ano', inplace=True)
def format_brazilian(x):
    return f'{x:,.2f}'.replace(',', 'X').replace('.', ',').replace('X', '.')
df_CCL['Capital Circulante Líquido'] = df_CCL['Capital Circulante Líquido'].apply(format_brazilian)

df_CCL
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Giro do Ativo}=\frac{\text{Receita de Vendas}}{\text{Ativo Total Médio}}$$</span>

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Endividamento}=\frac{\text{Passivo Total}}{\text{Patrimônio Líquido}}$$</span>


```python
#Endividamento em 2024
endv_2024 = (PC_2024+PNC_2024)/PL_2024

#Endividamento em 2023
endv_2023 = (PC_2023+PNC_2023)/PL_2023

#Endividamento em 2022
endv_2022 = (PC_2022+PNC_2022)/PL_2022

#Endividamento em 2021
endv_2021 = (PC_2021+PNC_2021)/PL_2021

#Endividamento em 2020
endv_2020 = (PC_2020+PNC_2020)/PL_2020

#DataFrame com os resultados
endividamento = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Endividamento': [endv_2020, endv_2021, endv_2022, endv_2023, endv_2024]
}

df_endividamento = pd.DataFrame(endividamento)

#Definindo o ano como índice
df_endividamento.set_index('Ano', inplace=True)

df_endividamento
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Composição da Dívida}=\frac{\text{Passivo Circulante}}{\text{Passivo Total}}$$</span>


```python
#Composição do Endividamento em 2024
div_2024 = PC_2024/(PC_2024+PNC_2024)

#Composição do Endividamento em 2023
div_2023 = PC_2023/(PC_2023+PNC_2023)

#Composição do Endividamento em 2022
div_2022 = PC_2022/(PC_2022+PNC_2022)

#Composição do Endividamento em 2021
div_2021 = PC_2021/(PC_2021+PNC_2021)

#Composição do Endividamento em 2020
div_2020 = PC_2020/(PC_2020+PNC_2020)

#DataFrame com os resultados
div = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Composição da Dívida': [div_2020, div_2021, div_2022, div_2023, div_2024]
}

df_div = pd.DataFrame(div)

#Definindo o ano como índice
df_div.set_index('Ano', inplace=True)

df_div
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$\text{Dependência Financeira}=\frac{\text{Passivo Total}}{\text{Ativo Total}}$$</span>


```python
#Dependência em 2024
dep_2024 = (PC_2024+PNC_2024)/(AC_2024+ANC_2024)

#Dependência em 2023
dep_2023 = (PC_2023+PNC_2023)/(AC_2023+ANC_2023)

#Dependência em 2022
dep_2022 = (PC_2022+PNC_2022)/(AC_2022+ANC_2022)

#Dependência em 2021
dep_2021 = (PC_2021+PNC_2021)/(AC_2021+ANC_2021)

#Dependência em 2023
dep_2020 = (PC_2020+PNC_2020)/(AC_2020+ANC_2020)

#DataFrame com os resultados
dep = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Dependência Financeira': [dep_2020, dep_2021, dep_2022, dep_2023, dep_2024]
}

df_dep = pd.DataFrame(dep)

#Definindo o ano como índice
df_dep.set_index('Ano', inplace=True)

df_dep
```

<span style="font-size: 28px; color: Black; font-family: 'Times New Roman';"> **5. Indicadores de Falência**</span>

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$Z=0.012\left(\frac{CCL}{Ativo\ Total}\right)+0.014\ \left(\frac{Lucros\ Retidos}{Ativo\ Total}\right)+0.033\left(\frac{EBIT}{Ativo\ Total}\right)+0.006\left(\frac{\text{Valor de Mercado da Ações}}{\text{Exigível}}\right)+0.999\left(\frac{\text{Vendas}}{\text{Ativo Total}}\right)$$</span>

```python
#Teste Z de Altman em 2024
Z_2024 = (0.012*(CCL_2024/(ANC_2024+AC_2024))+
          0.014*(LR_2024/(AC_2024+ANC_2024))+
          0.033*(EBIT_2024/(ANC_2024+AC_2024))+
          0.006*((12.76*839505097)/(PNC_2024+PC_2024))+
          0.999*(REC_2024/(AC_2024+ANC_2024)))

#Teste Z de Altman em 2023
Z_2023 = (0.012*(CCL_2023/(ANC_2023+AC_2023))+
          0.014*(LR_2023/(AC_2023+ANC_2023))+
          0.033*(EBIT_2023/(ANC_2023+AC_2023))+
          0.006*((16.21*839505097)/(PNC_2023+PC_2023))+
          0.999*(REC_2023/(AC_2023+ANC_2023)))

#Teste Z de Altman em 2022
Z_2022 = (0.012*(CCL_2022/(ANC_2022+AC_2022))+
          0.014*(LR_2022/(AC_2022+ANC_2022))+
          0.033*(EBIT_2022/(ANC_2022+AC_2022))+
          0.006*((11.15*839505097)/(PNC_2022+PC_2022))+
          0.999*(REC_2022/(AC_2023+ANC_2022)))

#Teste Z de Altman em 2021
Z_2021 = (0.012*(CCL_2021/(ANC_2021+AC_2021))+
          0.014*(LR_2021/(AC_2021+ANC_2021))+
          0.033*(EBIT_2021/(ANC_2021+AC_2021))+
          0.006*((24.26*839505097)/(PNC_2021+PC_2021))+
          0.999*(REC_2021/(AC_2021+ANC_2021)))

#Teste Z de Altman em 2020
Z_2020 = (0.012*(CCL_2020/(ANC_2020+AC_2020))+
          0.014*(LR_2020/(AC_2020+ANC_2020))+
          0.033*(EBIT_2020/(ANC_2020+AC_2020))+
          0.006*((50.09*839505097)/(PNC_2020+PC_2020))+
          0.999*(REC_2020/(AC_2023+ANC_2020)))

#DataFrame com os resultados
Z = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Teste Z de Altman': [Z_2020, Z_2021, Z_2022, Z_2023, Z_2024]
}

df_Z = pd.DataFrame(Z)

#Definindo o ano como índice
df_Z.set_index('Ano', inplace=True)

df_Z
```

<span style="font-size: 18px; color: Black; font-family: 'Times New Roman';"> $$F=0.05\left(\frac{\text{Lucro Líquido}}{\text{PL}}\right)+1.65\left(\frac{\mathrm{AC}+\mathrm{Realizavel\ a\ LP}}{\text{Exigível}}\right)+3.55\left(\frac{\mathrm{AC}-\mathrm{Estoques}}{\text{Exigível a CP}}\right)-1.06\left(\frac{\mathrm{AC}}{\text{Exigível a CP}}\right)-0.33\left(\frac{\text{Exigível}}{\mathrm{PL}}\right)$$</span>

```python
#Termômetro de Kanitz em 2024
F_2024 = (0.05*(RECL_2024/PL_2024)+
          1.65*((AC_2024+RLP_2024)/(PNC_2024+PC_2024))+
          3.55*((AC_2024-EST_2024)/PC_2024)-
          1.06*(AC_2024/PC_2024)-
          0.33*((PC_2024+PNC_2024)/PL_2024))

#Termômetro de Kanitz em 2023
F_2023 = (0.05*(RECL_2023/PL_2023)+
          1.65*((AC_2023+RLP_2023)/(PNC_2023+PC_2023))+
          3.55*((AC_2023-EST_2023)/PC_2023)-
          1.06*(AC_2023/PC_2023)-
          0.33*((PC_2023+PNC_2023)/PL_2023))

#Termômetro de Kanitz em 2022
F_2022 = (0.05*(RECL_2022/PL_2022)+
          1.65*((AC_2022+RLP_2022)/(PNC_2022+PC_2022))+
          3.55*((AC_2022-EST_2022)/PC_2022)-
          1.06*(AC_2022/PC_2022)-
          0.33*((PC_2022+PNC_2022)/PL_2022))

#Termômetro de Kanitz em 2021
F_2021 = (0.05*(RECL_2021/PL_2021)+
          1.65*((AC_2021+RLP_2021)/(PNC_2021+PC_2021))+
          3.55*((AC_2021-EST_2021)/PC_2021)-
          1.06*(AC_2021/PC_2021)-
          0.33*((PC_2021+PNC_2021)/PL_2021))

#Termômetro de Kanitz em 2020
F_2020 = (0.05*(RECL_2020/PL_2020)+
          1.65*((AC_2020+RLP_2020)/(PNC_2020+PC_2020))+
          3.55*((AC_2020-EST_2020)/PC_2020)-
          1.06*(AC_2020/PC_2020)-
          0.33*((PC_2020+PNC_2020)/PL_2020))

#DataFrame com os resultados
F = {
    'Ano': [2020, 2021, 2022, 2023, 2024],
    'Termômetro de Kanitz': [F_2020, F_2021, F_2022, F_2023, F_2024]
}

df_F = pd.DataFrame(F)

#Definindo o ano como índice
df_F.set_index('Ano', inplace=True)

df_F
```
